```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Netflix Userbase.csv")
df
```

```python
average_revenue = df.groupby("Subscription Type")["Monthly Revenue"].mean()
print(average_revenue)
```

```python
country = df.groupby("Country")["User ID"].count()
print(country)
```

```python
age = df.groupby("Plan Duration")["Age"].median()
print(age)
```

```python
gender = df.groupby("Gender")["Monthly Revenue"].sum()
print(gender)
```

```python
average_device_age = df.groupby("Device")["Age"].mean()
print(average_device_age)
```

```python
plt.figure(figsize =(10,6))
sns.scatterplot(x="Age", y="Monthly Revenue", hue="Gender",data=df)
plt.xlabel=("Age")
plt.ylabel=("Monthly Revenue")
plt.title("Age & Monthly Revenue")
plt.show()
```

```python
average_revenue_by_country = df.groupby("Country")["Monthly Revenue"].mean().reset_index()
sns.barplot(x="Country", y="Monthly Revenue", data=average_revenue_by_country)
plt.title("Revenue by Country")
plt.xlabel("Country")
plt.ylabel("Average Monthly Revenue")
plt.show()
```

```python
sns.countplot(x='Subscription Type', data = df, color = 'green')
plt.title('Subscription Type Distribution')
plt.xlabel('Subscription Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df["Monthly Revenue"], bins=5,kde=True)
plt.title('Monthly revenue Distribution')
plt.xlabel('Revenue Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df["Age"], bins=5,kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
